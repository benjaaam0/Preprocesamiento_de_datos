## Operaciones para Variables Cualitativas (Nominales)


```python
# Variables involucradas: Name, Nationality, Sport, Previous Year Rank.

# Con la variable Previous Year Rank es dificil realizar alguna operación, debido a que tiene valores >30, haciendo mas dificil hacer operaciones matematicas.

# Con el resto de Variables se puede hacer operaciones similares como con Sport, como la moda con Name, Nationality, y un conteo total de cada Pais y nombre que se repite.

# Variable Sport
datos_cat = df['Sport']

# Frecuencias
frecuencia_absoluta = datos_cat.value_counts()
frecuencia_relativa = datos_cat.value_counts(normalize=True) * 100

print("Frecuencia Absoluta (conteo):\n", frecuencia_absoluta.head())
print("Frecuencia Relativa (%):\n", frecuencia_relativa.head())

# Medida de Tendencia Central
print(f"Moda: {datos_cat.mode()[0]}")

# Gráfico
frecuencia_absoluta.head(10).plot(kind='bar')
plt.title('Distribución por Deporte')
plt.show()
```

## Operaciones para Variables Cuantitativas (Numéricas)

```python
# Variable involucrada: earnings ($ million)

datos_num = df['earnings ($ million)']


# Medidas de Tendencia Central
print(f"Media: {datos_num.mean()}")
print(f"Mediana: {datos_num.median()}")
print(f"Moda: {datos_num.mode()[0]}")

# Medidas de Dispersión
print(f"Varianza: {datos_num.var()}")
print(f"Desviación Estándar: {datos_num.std()}")
print(f"Rango: {datos_num.max() - datos_num.min()}")

# Coeficiente de Variación (CV = std / media)
cv = (datos_num.std() / datos_num.mean()) * 100
print(f"Coeficiente de Variación: {cv}%")

# Medidas de Posición (Cuartiles)
print("Cuartiles (Q1, Q2, Q3):")
print(datos_num.quantile( [0.25, 0.50, 0.75] )) # Pista: .quantile()

# Gráfico (Cálculo diferencial: Tendencia)
plt.plot(df['Year'], df['earnings ($ million)'], 'o', alpha=0.5)
plt.title('Dispersión de Ganancias por Año')
plt.show()
```
