## Operaciones para Variables Cualitativas (Nominales)


```python
import matplotlib.pyplot as plt

# =========================================================
# ANÁLISIS DE LA VARIABLE: SPORT
# =========================================================

# 1. Selección de la variable
datos_cat = df['Sport']

# 2. Cálculo de Frecuencias
frecuencia_absoluta = datos_cat.value_counts()
frecuencia_relativa = datos_cat.value_counts(normalize=True) * 100

# 3. Impresión de Resultados
print("=" * 45)
print(" RESULTADOS: VARIABLE 'SPORT'")
print("=" * 45)

print("\n🔹 FRECUENCIA ABSOLUTA (Top 5 Deportes):")
print(frecuencia_absoluta.head().to_string())

print("\n🔹 FRECUENCIA RELATIVA (%):")
# Formateamos a 2 decimales para que se vea más limpio
print(frecuencia_relativa.head().round(2).to_string())

print("\n🔹 MEDIDA DE TENDENCIA CENTRAL:")
print(f"Moda (Deporte más frecuente): {datos_cat.mode()[0]}")
print("=" * 45)

# 4. Visualización de Datos (Gráfico de Barras)
# Ajustamos un poco el tamaño y color para que se vea mejor
plt.figure(figsize=(10, 6))
frecuencia_absoluta.head(10).plot(kind='bar', color='skyblue', edgecolor='black')

# Etiquetas del gráfico
plt.title('Top 10 - Distribución por Deporte', fontsize=14, fontweight='bold')
plt.xlabel('Deporte', fontsize=12)
plt.ylabel('Cantidad de Atletas', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rota los nombres para que no se amontonen

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

## Operaciones para Variables Cuantitativas (Numéricas)

```python
import matplotlib.pyplot as plt

# =========================================================
# ANÁLISIS DE LA VARIABLE: EARNINGS ($ MILLION)
# =========================================================

datos_num = df['earnings ($ million)']

print("=" * 55)
print(" 📊 RESULTADOS: VARIABLE 'EARNINGS ($ MILLION)'")
print("=" * 55)

# 1. Medidas de Tendencia Central
print("\n🔹 MEDIDAS DE TENDENCIA CENTRAL:")
print(f"   - Media (Promedio)    : {datos_num.mean():.2f}")
print(f"   - Mediana             : {datos_num.median():.2f}")
print(f"   - Moda                : {datos_num.mode()[0]:.2f}")

# 2. Medidas de Dispersión
print("\n🔹 MEDIDAS DE DISPERSIÓN:")
print(f"   - Varianza            : {datos_num.var():.2f}")
print(f"   - Desviación Estándar : {datos_num.std():.2f}")
print(f"   - Rango Estadístico   : {(datos_num.max() - datos_num.min()):.2f}")

# 3. Coeficiente de Variación (CV = std / media * 100)
cv = (datos_num.std() / datos_num.mean()) * 100
print(f"   - Coef. de Variación  : {cv:.2f}%")

# 4. Medidas de Posición (Cuartiles)
print("\n🔹 MEDIDAS DE POSICIÓN (CUARTILES):")
cuartiles = datos_num.quantile([0.25, 0.50, 0.75])
print(f"   - Q1 (25%)            : {cuartiles[0.25]:.2f}")
print(f"   - Q2 (50% / Mediana)  : {cuartiles[0.50]:.2f}")
print(f"   - Q3 (75%)            : {cuartiles[0.75]:.2f}")
print("=" * 55)

# 5. Visualización: Gráfico de Dispersión (Ganancias vs Año)
plt.figure(figsize=(10, 6))

# Dibujamos el gráfico de dispersión (alpha da transparencia para ver puntos superpuestos)
plt.plot(df['Year'], datos_num, 'o', color='coral', alpha=0.6, markersize=7)

# Etiquetas y personalización del gráfico
plt.title('Dispersión y Tendencia de Ganancias por Año', fontsize=14, fontweight='bold')
plt.xlabel('Año', fontsize=12)
plt.ylabel('Ganancias ($ millones)', fontsize=12)

# Agregamos una cuadrícula de fondo para facilitar la lectura
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar el gráfico ajustado
plt.tight_layout()
plt.show()
```
